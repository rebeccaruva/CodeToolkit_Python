### Shephard Fairey

```python
rm = io.imread("rmGlasses.png")
```

```python
rm.shape
```

```python
plt.imshow(rm)
```

```python
grayscale = rgb2gray(rm)
```

```python
plt.imshow(grayscale, cmap=plt.cm.gray)
```

```python
grayscale.shape
```

```python
grayscale[4, 9]
```

```python
rmboom = np.zeros((666, 512, 3))

for i in range(666):
    for j in range(512):
        if grayscale[i, j] > 0.75:
            rmboom[i, j] = [160, 65, 220]
        elif grayscale[i, j] > 0.5:
            rmboom[i, j] = [165, 165, 165]
        elif grayscale[i, j] > 0.25:
            rmboom[i, j] = [65, 155, 230]
        else:
            rmboom[i, j] = [255, 200, 240]
```

```python
rmboom2 = np.zeros((666, 512, 3))

for i in range(666):
    for j in range(512):
        if grayscale[i, j] > 0.75:
            rmboom2[i, j] = [0, 0, 255]
        elif grayscale[i, j] > 0.5:
            rmboom2[i, j] = [0, 255, 0]
        elif grayscale[i, j] > 0.25:
            rmboom2[i, j] = [20, 20, 20]
        else:
            rmboom2[i, j] = [0, 0, 0]
```

```python
plt.imshow(rmboom2)
```

```python
plt.imshow(rmboom)
```

```python
def posterizeIt(image_file):
    p = io.imread(image_file)
    grayscale = rgb2gray(p)
    rows = grayscale.shape[0]
    cols = grayscale.shape[1]
    rmboom = np.zeros((rows, cols, 3))
    for i in range(rows):
        for j in range(cols):
            if grayscale[i, j] > 0.75:
                rmboom[i, j] = [255, 0, 255]
            elif grayscale[i, j] > 0.5:
                rmboom[i, j] = [255, 255, 0]
            elif grayscale[i, j] > 0.25:
                rmboom[i, j] = [20, 20, 20]
            else:
                rmboom[i, j] = [0, 0, 0]
    return plt.imshow(rmboom)
```

```python
posterizeIt("rmGlasses.png")
```

```python
posterizeIt("enhance.computer.PNG")
```

```python
posterizeIt("rm1.jpg")
```

```python
posterizeIt("rm2.jpg")
```

```python
posterizeIt("rm3.jpg")
```
